## Viaticos

Total gastado por año

```python
%sql
SELECT
    t.anio,
    SUM(f.monto)    AS total_soles,
    COUNT(*)        AS num_registros
FROM corpac_viaticos.fact_viaticos f
JOIN corpac_viaticos.dim_tiempo t ON f.id_tiempo = t.id_tiempo
GROUP BY t.anio
ORDER BY t.anio;
```

Total por rubro (ranking general)

```python
%sql
SELECT
    r.rubro,
    SUM(f.monto)    AS total_soles,
    COUNT(*)        AS num_registros
FROM corpac_viaticos.fact_viaticos f
JOIN corpac_viaticos.dim_rubro  r ON f.id_rubro  = r.id_rubro
GROUP BY r.rubro
ORDER BY total_soles DESC;
```

## Penalidades

Total de penalidades por año

```python
%sql
SELECT
    t.anio,
    COUNT(*)                        AS num_penalidades,
    SUM(f.monto_penalidad)          AS total_penalidad,
    SUM(f.monto_total_contrato)     AS total_contrato,
    ROUND(
        SUM(f.monto_penalidad) / NULLIF(SUM(f.monto_total_contrato), 0) * 100, 2
    )                               AS pct_penalidad_sobre_contrato
FROM corpac_penalidades.fact_penalidad f
JOIN corpac_penalidades.dim_tiempo     t ON f.id_tiempo = t.id_tiempo
GROUP BY t.anio
ORDER BY t.anio;
```

Proveedores con más penalidades (ranking)

```python
%sql
SELECT
    p.proveedor,
    p.ruc,
    COUNT(*)                    AS veces_penalizado,
    SUM(f.monto_penalidad)      AS total_penalidad
FROM corpac_penalidades.fact_penalidad f
JOIN corpac_penalidades.dim_proveedor  p ON f.id_proveedor = p.id_proveedor
GROUP BY p.proveedor, p.ruc
ORDER BY total_penalidad DESC
LIMIT 20;
```

## ORDENES

Total de órdenes por año

```python
%sql
SELECT
    t.anio,
    COUNT(*)            AS num_ordenes,
    SUM(f.monto_soles)  AS total_soles
FROM corpac_ordenes.fact_orden f
JOIN corpac_ordenes.dim_tiempo t ON f.id_tiempo = t.id_tiempo
GROUP BY t.anio
ORDER BY t.anio
```

Total por origen (Sede Central vs Provincias)

```python
%sql
SELECT
    o.origen,
    t.anio,
    COUNT(*)            AS num_ordenes,
    SUM(f.monto_soles)  AS total_soles
FROM corpac_ordenes.fact_orden   f
JOIN corpac_ordenes.dim_tiempo   t ON f.id_tiempo = t.id_tiempo
JOIN corpac_ordenes.dim_origen   o ON f.id_origen = o.id_origen
GROUP BY o.origen, t.anio
ORDER BY t.anio, total_soles DESC;
```

Proveedores con mayor monto contratado (ranking)

```python
%sql
SELECT
    p.proveedor,
    p.ruc,
    COUNT(*)            AS num_ordenes,
    SUM(f.monto_soles)  AS total_soles
FROM corpac_ordenes.fact_orden    f
JOIN corpac_ordenes.dim_proveedor p ON f.id_proveedor = p.id_proveedor
GROUP BY p.proveedor, p.ruc
ORDER BY total_soles DESC
LIMIT 20;
 
```
